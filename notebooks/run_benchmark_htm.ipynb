## Get anomaly windows from the files

```python

from orion.primitives.timeseries_anomalies import find_anomalies
import pandas as pd
import numpy as np
from IPython.display import display
from orion.data import load_anomalies
import math
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

from datetime import datetime 
def str_to_timestamp(s):
    return datetime.strptime(s, '%Y-%m-%d %H:%M:%S').timestamp()

def get_anomalies_from_results_1(signal):
    """ Use find_anomaly primitive """
    
    df = pd.read_csv( './results_htm/{}.csv'.format(signal))
    df['timestamp'] = df['timestamp'].apply(str_to_timestamp)
#     print(df.head())
#     print(df.info())
#     print(df.shape[0])
    errors = df['anomaly_score'].to_numpy()
    error_smooth_window = math.trunc(len(errors) * 0.01)
    errors = pd.Series(errors).rolling(
        error_smooth_window, center=True, min_periods=error_smooth_window//2).mean().values
    
    index = df['timestamp'].astype('int64').to_numpy()
    anomalies = find_anomalies(errors, index)
    events = list()
    for anomaly in anomalies:
        events.append(anomaly)
    if len(events) > 0:
        events = pd.DataFrame(np.vstack(events), columns=['start', 'end', 'score'])
        events['start'] = events['start'].astype(int)
        events['end'] = events['end'].astype(int)
    return events

def get_anomalies_from_results_2(signal, th=None):
    """ Manually specify threshold
        By default, use mean + 4std
    """
    
    df = pd.read_csv( './results_htm/{}.csv'.format(signal))
    df['timestamp'] = df['timestamp'].apply(str_to_timestamp)
    length = df.shape[0]
    st = 0
    anomaly_num = 0
    anomalies = pd.DataFrame([], columns = ['start', 'end'])
    if th is None:
        th = df['anomaly_score'].mean() + 4 * df['anomaly_score'].std()
    while st < length:
        if df.iloc[st]['anomaly_score'] < th:
            st += 1
            continue
        ed = st
        while (ed + 1 < length) and (df.iloc[ed + 1]['anomaly_score'] >= th):
            ed += 1
        anomalies.loc[anomaly_num] = {'start': df.iloc[st]['timestamp'], 'end': df.iloc[ed]['timestamp']}
        anomaly_num += 1
        st = ed + 1
    return anomalies

def get_anomalies_from_results_3(signal, th=1):
    """ Use label: current data do not have labels info """
    
    df = pd.read_csv( './results_htm/{}.csv'.format(signal))
    df['timestamp'] = df['timestamp'].apply(str_to_timestamp)
#     print(df.info())
    
    length = df.shape[0]
    st = 0
    anomaly_num = 0
    anomalies = pd.DataFrame([], columns = ['start', 'end'])
    while st < length:
        if df.iloc[st]['label'] < 1:
            st += 1
            continue
        ed = st
        while (ed + 1 < length) and (df.iloc[ed + 1]['label'] >= 1):
            ed += 1
        anomalies.loc[anomaly_num] = {'start': df.iloc[st]['timestamp'], 'end': df.iloc[ed]['timestamp']}
        anomaly_num += 1
        st = ed + 1
    return anomalies

get_anomalies_from_results_1('M-6')

```

## Compute scores

```python
from orion.data import load_anomalies
from orion import metrics
from orion.metrics import score_overlap
warnings.simplefilter(action='ignore', category=FutureWarning)

METRICS = {
    'accuracy': metrics.accuracy_score,
    'precision': metrics.precision_score,
    'recall': metrics.recall_score,
    'f1': metrics.f1_score,
}

signals_set = ALL
name_set = ALL_NAME

scores_set = list()
bad_num = 0

for i, signals in enumerate(signals_set):
    scores_weighted_list = list()
    tp_sum, fp_sum, fn_sum = 0, 0, 0

    for signal in signals:
        truth = load_anomalies(signal)
        try:
            anomalies = get_anomalies_from_results_1(signal)
#             scores_weighted_item = {
#                 name: scorer(truth, anomalies)
#                 for name, scorer in METRICS.items()
#             }
            tp, fp, fn = score_overlap(truth, anomalies)
        except Exception:
            bad_num += 1
#             scores_weighted_item = {name: 0 for name in METRICS.keys()}
            tp, fp, fn = 0, 0, 0
        
        tp_sum += tp
        fp_sum += fp
        fn_sum += fn
#         scores_weighted_list.append(scores_weighted_item)
    
#     scores_weighted_df = pd.DataFrame(scores_weighted_list).mean()
#     print(scores_weighted_df)
    
    # manually compute the score for one dataset
    precision = tp_sum / (tp_sum + fp_sum)
    recall = tp_sum / (tp_sum + fn_sum)
    f1 = 2 * precision * recall / (precision + recall)
    scores = {
        'tp': tp_sum,
        'fp': fp_sum,
        'fn': fn_sum,
        'np': precision,
        'nr': recall,
        'nf': f1
    }
    print(name_set[i])
    print(scores)
    scores_set.append(scores)


```
