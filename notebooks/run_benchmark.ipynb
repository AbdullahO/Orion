```python
import keras
import pandas as pd
from orion.evaluation import evaluate_pipelines 
from orion.evaluation_variants import evaluate_pipelines as evaluate_pipelines_variants 
from IPython.display import display
import warnings
warnings.filterwarnings("ignore")

"""
ONLY FOR TESTING
Choose subset of signals
Please comment the following lines when producing full results
"""
SMAP = SMAP[:1]
MSL = MSL[:1]
YAHOOA1_SIGNALS = YAHOOA1_SIGNALS[:1]
YAHOOA2_SIGNALS = YAHOOA2_SIGNALS[:1]
YAHOOA3_SIGNALS = YAHOOA3_SIGNALS[:1]
YAHOOA4_SIGNALS = YAHOOA4_SIGNALS[:1]
artificialWithAnomaly = artificialWithAnomaly[:1]
realAdExchange = realAdExchange[:1]
realAWSCloudwatch = realAWSCloudwatch[:1]
realTraffic = realTraffic[:1]
realTweets = realTweets[:1]

def compute_scores(row):
    (tp, fp, fn) = (row['tp'], row['fp'], row['fn'])
    try:
        p = tp / (tp + fp)
        r = tp / (tp + fn)
        f1 = 2 * (p * r) / (p + r)
    except:
        p, r, f1 = 0, 0, 0
    return (f1, p, r)


def get_scores(pipelines, signals, **kargs):
    nf, np, nr = [], [], []
    scores = evaluate_pipelines(pipelines=pipelines, signals=signals, **kargs)
    
    for index, row in scores.iterrows():
        (f,p,r) = compute_scores(row)
        nf.append(f)
        np.append(p)
        nr.append(r)
    scores['nf'] = nf
    scores['np'] = np
    scores['nr'] = nr
    return scores

def get_scores_variants(pipelines, signals, name, **kargs):
    nf, np, nr = [], [], []
    scores = evaluate_pipelines_variants(pipelines=pipelines, signals=signals, **kargs)
    
    dts = list()
    for pipeline_score in scores:
        dt = {
            'acc': [],
            'f1': [],
            'pre': [],
            'rec': [],
            'tp': [],
            'fp': [],
            'fn': [],
            'nf1': [],
            'npre': [],
            'nrec': [],
            'pipeline': [],
            'dataset': [name] * len(pipeline_score),
        }
        
        for s in pipeline_score:
            dt['acc'].append(s['accuracy'])
            dt['f1'].append(s['f1'])
            dt['pre'].append(s['precision'])
            dt['rec'].append(s['recall'])
            dt['tp'].append(s['tp'])
            dt['fp'].append(s['fp'])
            dt['fn'].append(s['fn'])

            (f, p, r) = compute_scores(s)
            dt['nf1'].append(f)
            dt['npre'].append(p)
            dt['nrec'].append(r)
            
            dt['pipeline'].append(s['pipeline'].replace('../orion/pipelines/', ''))
        dt = pd.DataFrame(dt)
        dts.append(dt)
    return dts 

def print_scores(scores):
    if (type(scores) is list):
        for dt in scores:
            display(dt)
    else:
        for index, row in scores.iterrows():
            row['pipeline'] = row['pipeline'].replace('../orion/pipelines/', '')
            scores.loc[index] = row
        scores = scores.drop(['holdout', 'elapsed'], axis=1)
        scores = scores.sort_values(by=['pipeline'])
        display(scores)
        print('*************************************')

scores_set = list()

# """
# Non-cgan pipelines
# """
# pipelines = [
#     '../orion/pipelines/lstm_dynamic_threshold_gpu_yahoo.json'
# ]
# scores = get_scores(pipelines, YAHOOA1_SIGNALS, detrend=True, split=True, holdout=False)
# print_scores(scores)
# scores_set.append(scores)
# scores = get_scores(pipelines, YAHOOA2_SIGNALS, detrend=True, split=True, holdout=False)
# print_scores(scores)
# scores_set.append(scores)
# scores = get_scores(pipelines, YAHOOA3_SIGNALS, detrend=True, split=True, holdout=False)
# print_scores(scores)
# scores_set.append(scores)
# scores = get_scores(pipelines, YAHOOA4_SIGNALS, detrend=True, split=True, holdout=False)
# print_scores(scores)
# scores_set.append(scores)

# pipelines = [
#     '../orion/pipelines/lstm_dynamic_threshold_gpu_nab.json'
# ]
# scores = get_scores(pipelines, artificialWithAnomaly, split=True, holdout=False)
# print_scores(scores)
# scores_set.append(scores)
# scores = get_scores(pipelines, realAdExchange, split=True, holdout=False)
# print_scores(scores)
# scores_set.append(scores)
# scores = get_scores(pipelines, realAWSCloudwatch, split=True, holdout=False)
# print_scores(scores)
# scores_set.append(scores)
# scores = get_scores(pipelines, realTraffic, split=True, holdout=False)
# print_scores(scores)
# scores_set.append(scores)
# scores = get_scores(pipelines, realTweets, split=True, holdout=False)
# print_scores(scores)
# scores_set.append(scores)


# pipelines = [
#     '../orion/pipelines/lstm_dynamic_threshold_gpu_nasa.json'
# ]
# scores = get_scores(pipelines, MSL, holdout=True)
# print_scores(scores)
# scores_set.append(scores)
# scores = get_scores(pipelines, SMAP, holdout=True)
# print_scores(scores)
# scores_set.append(scores)

"""
TadGAN pipelines
"""
######## yahoo
pipelines = [
    '../orion/pipelines/tadgan_variants_yahoo.json'
]

scores = get_scores_variants(pipelines, YAHOOA1_SIGNALS, name="A1", detrend=True, split=True, holdout=False)
print_scores(scores)
scores_set.append(scores)

scores = get_scores_variants(pipelines, YAHOOA2_SIGNALS, name="A2", detrend=True, split=True, holdout=False)
print_scores(scores)
scores_set.append(scores)

scores = get_scores_variants(pipelines, YAHOOA3_SIGNALS, name="A3", detrend=True, split=True, holdout=False)
print_scores(scores)
scores_set.append(scores)

scores = get_scores_variants(pipelines, YAHOOA4_SIGNALS, name="A4", detrend=True, split=True, holdout=False)
print_scores(scores)
scores_set.append(scores)

######## nab
pipelines = [
    '../orion/pipelines/tadgan_variants_nab.json'
]

scores = get_scores_variants(pipelines, artificialWithAnomaly, name="art", holdout=False, split=True)
print_scores(scores)
scores_set.append(scores)

scores = get_scores_variants(pipelines, realAdExchange, name="ade", holdout=False, split=True)
print_scores(scores)
scores_set.append(scores)

scores = get_scores_variants(pipelines, realAWSCloudwatch, name="aws", holdout=False, split=True)
print_scores(scores)
scores_set.append(scores)

scores = get_scores_variants(pipelines, realTraffic, name="traffic", holdout=False, split=True)
print_scores(scores)
scores_set.append(scores)

scores = get_scores_variants(pipelines, realTweets, name="tweets", holdout=False, split=True)
print_scores(scores)
scores_set.append(scores)

######## nasa
pipelines = [
    '../orion/pipelines/tadgan_variants_nasa.json'
]

scores = get_scores_variants(pipelines, MSL, name="msl", holdout=True)
print_scores(scores)
scores_set.append(scores)

scores = get_scores_variants(pipelines, SMAP, name="smap", holdout=True)
print_scores(scores)
scores_set.append(scores)

```
